Forecasting with FB prophet algorithm

```python
#converting the date from DD/MM/YYYY to YYYY-MM-DD

date_info = pd.to_datetime(df['Date'], dayfirst=True)
print(df['Date'])
```

```python
time_info = df['Time']
```

```python
time_info = time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
#combining 2 series to a panda dataframe
date_time = pd.concat([date_info , time_info] , axis=1)
```

```python
date_time.head()
```

```python
#combining date and time

date_time['ds'] = date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
```

```python
date_time.info()
```

we have to convert ds from object data type to datetime format

```python
data =pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

we will make future predictions for relative humidity

```python
data['y'] = df['RH']
```

```python
data.head()
```

```python
import sys
!{sys.executable} -m pip install prophet
```

```python
from prophet import Prophet
```

```python
#training the prophet model

model = Prophet()
model.fit(data)
```

```python
#to make future predictions
future = model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds' , 'yhat' , 'yhat_lower' , 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```
